```python
from matplotlib import pyplot
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('molnet_bace_regression_models_prediction_results.csv')
```

```python
fig=plt.figure(figsize=(15,10))
ax1=fig.add_subplot(2,2,1)
pyplot.scatter(data1['true_pIC50'], data1['xgb_test_pIC50'])
pyplot.plot(np.arange(3, 10), np.arange(3, 10), color='red')
pyplot.xlabel('true_pIC50')
pyplot.ylabel('xgb_test_pIC50')
pyplot.title('XGBoost')
pyplot.xlim(3, 9.5)
pyplot.ylim(3, 9.5)
ax1.legend(['R2=0.65', 'R2=1'])

ax2=fig.add_subplot(2,2,2)
pyplot.scatter(data1['true_pIC50'], data1['svm_test_pIC50'])
pyplot.plot(np.arange(3, 10), np.arange(3, 10), color='red')
pyplot.xlabel('true_pIC50')
pyplot.ylabel('svm_test_pIC50')
pyplot.title('SVM')
pyplot.xlim(3, 9.5)
pyplot.ylim(3, 9.5)
ax2.legend(['R2=0.74', 'R2=1'])

ax3=fig.add_subplot(2,2,3)
pyplot.scatter(data1['true_pIC50'], data1['rf_test_pIC50'])
pyplot.plot(np.arange(3, 10), np.arange(3, 10), color='red')
pyplot.xlabel('true_pIC50')
pyplot.ylabel('rf_test_pIC50')
pyplot.title('Random Forest')
pyplot.xlim(3, 9.5)
pyplot.ylim(3, 9.5)
ax3.legend(['R2=0.66', 'R2=1'])

ax4=fig.add_subplot(2,2,4)
pyplot.scatter(data1['true_pIC50'], data1['knn_test_pIC50'])
pyplot.plot(np.arange(3, 10), np.arange(3, 10), color='red')
pyplot.xlabel('true_pIC50')
pyplot.ylabel('knn_test_pIC50')
pyplot.title('KNN')
pyplot.xlim(3, 9.5)
pyplot.ylim(3, 9.5)
ax4.legend(['R2=0.58', 'R2=1'])
```
